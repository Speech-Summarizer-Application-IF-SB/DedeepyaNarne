```python
import torch
from transformers import pipeline
import librosa
import soundfile as sf
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
file_path = "/content/drive/MyDrive/auido_mono.wav"
```

```python
asr_pipeline = pipeline(
    "automatic-speech-recognition",
    model="openai/whisper-small",
    device=0 if torch.cuda.is_available() else -1
)
```

```python
result = asr_pipeline(file_path)
print("Transcript:\n", result["text"])
```

```python
# Replace with your known ground truth text
ground_truth_text = """Hello everyone, welcome to the meeting. Today we'll discuss our project progress."""

# Compute Word Error Rate
error = wer(ground_truth_text.lower(), result.lower())
print(f"\n📊 Word Error Rate (WER): {error * 100:.2f}%")
```
